```python
#Time series and visualization packages
from dateutil.parser import parse 
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import numpy as np
import pandas as pd
plt.rcParams.update({'figure.figsize': (10, 7), 'figure.dpi': 120})

#possible forecasting through machine learning
import xgboost as xgb

#misc
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse
```

```python
fero = pd.read_csv("FOE.csv")
```

```python
#Taking a look at our data
fero.head()
```

```python
#numerical value summary statistics
fero[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']].describe()
```

```python
#plotting the closing price over the years
# Process dataset
fero['Date'] = pd.to_datetime(fero['Date'])
fero['Close'] = pd.to_numeric(fero['Close'])

# Plot
plt.figure(figsize=(16,5), dpi=120)
plt.plot(fero['Date'], fero['Close'])
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator(2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.title('FOE Closing prices since 1980', fontsize=14)
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.gcf().autofmt_xdate() # Rotation
plt.show()
```

```python
#Plot a time series graph for one year, we choose 2021
fero5y = fero.loc[(fero['Date'] >= '2017-01-01')]

plt.figure(figsize=(16,5), dpi=120)
plt.plot(fero5y['Date'], fero5y['Close'])
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator(interval = 6))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.title('FOE Closing prices from 2017 to 2022', fontsize=14)
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.gcf().autofmt_xdate() # Rotation
plt.show()
```

```python
#check for Stationarity
from statsmodels.tsa.stattools import adfuller, kpss
# ADF Test
adf = adfuller(fero5y["Close"], autolag='AIC') 
print(f'ADF Statistic: {adf [0]}')
print(f'p-value: {adf[1]}')
for key, value in adf[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')

# KPSS Test
kps = kpss(fero5y["Close"], regression='c')
print('\nKPSS Statistic: %f' % kps[0])
print('p-value: %f' % kps[1])
for key, value in kps[3].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
#the p-value of the adf statistic is more than 5%, we fail to the null hypothesis that the data has a unit root and is not stationary
#Alternatively the p-value of the KPSS test is much smaller that 5% providing evidence to the alternative hypothesis that the data is non-stationary
```

```python
#Time Series decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
decompose = seasonal_decompose(fero2021['Close'],model='additive', period=7)
decompose.plot()
plt.show()
#The seasonality plot seems to have a repeating pattern which could mean saesonality exists
```

```python
#Is Autocorrelation a problem? Durbin Watson Test performed
from statsmodels.stats.stattools import durbin_watson
dw = durbin_watson(fero2021['Close'])
print(dw)
#According to the Durbin Watson test, the acceptable range of values where Autocorrelation is not a problem is 1.5 to 2.5.
#Our value is outside this range (0.000303566832178246) therefore, autocorrelation has to be removed in order to perform Forecasting.
```

```python
#First differences
first_diff = fero2021['Close'].diff()[1:]
print(first_diff)
```

```python
#DW test on first diffferences
dw_fd = durbin_watson(first_diff)
print(dw_fd)
#The new DW value (2.110641391048694) is within the range of 1.5 to 2.5 thus autocorrelation is no longer problematic
```

```python
# ADF Test
adf2 = adfuller(first_diff, autolag='AIC') 
print(f'ADF Statistic: {adf2[0]}')
print(f'p-value: {adf2[1]}')
for key, value in adf2[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')

# KPSS Test
kps2 = kpss(first_diff, regression='c')
print('\nKPSS Statistic: %f' % kps2[0])
print('p-value: %f' % kps2[1])
for key, value in kps2[3].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
#Further evidence that data is now stationary, ADF test's p-value is now much lower than 0.05 thus the unit root no longer exists.
```
